# title

```python
from bs4 import BeautifulSoup
soup1 = BeautifulSoup(page_source,features="html.parser")
soup2=BeautifulSoup(page_source2,features="html.parser")
job_desc=[]
title1=soup1.find_all('div',class_="jobTitle")
title2=soup2.find_all('div',class_="jobTitle")
for i in title1[:15]:
    job_desc.append([i.text.strip()])
for i in title2[:15]:
    job_desc.append([i.text.strip()])
job_desc
```

# Location | Experience | Salary

```python
details=soup1.find_all('div',class_="cardBody")
details2=soup2.find_all('div',class_="cardBody")
m=0
for i in details[:]:
    k=i.find_all('div',class_="bodyRow")
    k=[j.text.strip() for j in i.find_all('div',class_="details") if "time" not in j.text]
    if "INR" not in k[-1]:
        k.append("Not Disclosed")
    
    if "Years" not in k[1] :
        k.insert(1,"0-0 Years")
    if len(k)==4:
        k.pop()
    for n in k:
        job_desc[m].append(n)
    m=m+1
for i in details2[:]:
    k=i.find_all('div',class_="bodyRow")
    k=[j.text.strip() for j in i.find_all('div',class_="details") if "time" not in j.text]
    if "INR" not in k[-1]:
        k.append("Not Disclosed")
    #print(k)
    if "Years" not in k[1] :
        k.insert(1,"0-0 Years")
    if len(k)==4:
        k.pop()
    #print(k)
    for n in k:
        job_desc[m].append(n)
    m=m+1

        
```

# skills

```python
card1=soup1.find_all('div',class_="cardContainer")
card2=soup2.find_all('div',class_="cardContainer")
k=0
for i in card1:
    skills=i.find_all('div',class_="skill")
    skills="/".join([i.text.lower().strip() for i in skills])
    if len(skills)!=0:
        job_desc[k].append(skills)
    else:
        job_desc[k].append("Not disclosed")
    k=k+1
for i in card2:
    skills=i.find_all('div',class_="skill")
    skills="/".join([i.text.lower().strip() for i in skills])
    if len(skills)!=0:
        job_desc[k].append(skills)
    else:
        job_desc[k].append("Not disclosed")
    k=k+1
    #print(skills)
    #job_desc.append([i.text.strip() for i in skills])
```

```python
min_years=[]
max_years=[]
for i in df["Work Experience"]:
    b=i.split()
    #print(b)
    if len(b)==2:
        c=[i for i in b[0].split('-') if i.isdigit()]
        if len(c)==0:
            min_years.append("Not Disclosed")
            max_years.append("Not Disclosed")
            
        else:
            #print(c)
            min_years.append(c[0])
            max_years.append(c[1])
    else:
        c=[i for i in b if i.isdigit()]
        min_years.append(c[0])
        max_years.append(c[1])
```

```python
df['min_years']=min_years
df['max_years']=max_years
```
