```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import *
import pandas as pd
import time
```

```python
driver=webdriver.Chrome(executable_path="C:\\BrowserDrivers\\chromedriver.exe")
driver.get("https://www.foundit.in/srp/results?query=%22Software+Developer%22&searchId=bb18f35f-b0ce-4cdd-a586-0a1dfdfb5ad7")
page_source=driver.page_source
```

```python
page_source
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page_source,features="html.parser")
```

#  Name

```python
soup = BeautifulSoup(page_source,features="html.parser")
job_desc=[]
title=soup.find_all('div',class_="jobTitle")
for i in title[:15]:
    job_desc.append([i.text.strip()])
job_desc
```

# Location | Experience | Salary

```python
details=soup.find_all('div',class_="cardBody")
m=0
for i in details[:]:
    k=i.find_all('div',class_="bodyRow")
    k=[j.text.strip() for j in i.find_all('div',class_="details") if "time" not in j.text]
    if "INR" not in k[-1]:
        k.append("Not Disclosed")
    #print(k)
    if "Years" not in k[1] :
        k.insert(1,"Not Disclosed")
    if len(k)==4:
        k.pop()
    #print(k)
    for n in k:
        job_desc[m].append(n)
    m=m+1

        
```

```python
job_desc
```

# Keyskill

```python
skillset=soup.find_all('div',class_="skillDetails")
k=0
for i in skillset:
    skills=i.find_all('div',class_="skill")
    a=[j.text.strip() for j in skills]
    job_desc[k].append(a)
    k=k+1
```

```python
job_desc
```

# Education

```python
for i in job_desc:
    i.append(['Any Graduate', 'Any Postgraduate'])
    
```

# Links

```python
driver=webdriver.Chrome(executable_path="C:\\BrowserDrivers\\chromedriver.exe")
driver.get("https://www.foundit.in/srp/results?query=%22Software+Developer%22&searchId=bb18f35f-b0ce-4cdd-a586-0a1dfdfb5ad7")
for i in range(15):
    job_desc[i].append(driver.current_url)
```
